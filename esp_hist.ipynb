```python
import cv2
import matplotlib.pyplot as plt
from funcoespi import esp_hist
```

```python
#carregar imagens
imagem_alvo = cv2.imread('data/flor_alvo.png')
imagem_referencia = cv2.imread('data/flor_referencia.png')

flor_resultado = esp_hist(imagem_referencia, imagem_alvo)

#exibir resultados

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.title('Imagem Alvo')
plt.imshow(cv2.cvtColor(imagem_alvo, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,3,2)
plt.title('Imagem Referência')
plt.imshow(cv2.cvtColor(imagem_referencia, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,3,3)
plt.title('Resultado ESP-Hist')
plt.imshow(cv2.cvtColor(flor_resultado, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
