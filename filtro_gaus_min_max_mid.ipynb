```python
import cv2
import matplotlib.pyplot as plt
from filtros import filtro_gaussiana
```

```python
#carregar imagens
flor_brg = cv2.imread('data/flor.png')
```

```python
flor_gauss = cv2.GaussianBlur(flor_brg, (5, 5), 0)
flor_gaussiana = filtro_gaussiana(flor_brg,10)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.title('Filtro Gaussiano OpenCV')
plt.imshow(cv2.cvtColor(flor_gauss, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,3,2)
plt.title('Imagem Original')
plt.imshow(cv2.cvtColor(flor_brg, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,3,3)
plt.title('Filtro Gaussiano Manual')
plt.imshow(flor_gaussiana)
plt.axis('off')
plt.show()
```
