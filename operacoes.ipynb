```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
# Carregar a imagem
imagem = cv2.imread('data/pirata.jpeg')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
plt.title('Original')
altura = imagem_rgb.shape[0]
largura = imagem_rgb.shape[1]
canais_cor = imagem_rgb.shape[2]
print (f'A largura da imagem é {largura}, a altura é {altura} e ela tem {canais_cor} canais de cor')
```

```python
#Transladar a imagem
tx, ty = 100, 50
M = np.float32([[1,0, tx],[0,1,ty]])
imagem_transladada = cv2.warpAffine(imagem_rgb, M, (largura, altura))
plt.subplot(1,2,1)
plt.imshow(imagem_rgb)
plt.title('Original')
plt.subplot(1,2,2)
plt.title('Transladada')
plt.imshow(imagem_transladada)
```

```python
# ZoomIn e ZoomOut

InSx, InSy = 1.5, 1.5
OutSx, OutSy = 0.5, 0.5

M_In = np.float32([[InSx,0,0],[0,InSy,0]])
M_Out = np.float32([[OutSx,0,0],[0,OutSy,0]])

imagem_zoom_in = cv2.warpAffine(imagem_rgb, M_In, (largura, altura))
imagem_zoom_out = cv2.warpAffine(imagem_rgb, M_Out, (largura, altura))
plt.subplot(1,2,1)
plt.imshow(imagem_zoom_in)
plt.title('Zoom in')
plt.subplot(1,2,2)
plt.imshow(imagem_zoom_out)
plt.title('Zoom out')
```

```python
#Zoom In e Zoom Out com cv2.resize
imagem_zoom_in = cv2.resize(imagem_rgb, None, fx=InSx, fy=InSy)
imagem_zoom_out = cv2.resize(imagem_rgb, None, fx=OutSx, fy=OutSy)
plt.subplot(1,2,1)
plt.imshow(imagem_zoom_in)
plt.title('Zoom in')
plt.subplot(1,2,2)
plt.imshow(imagem_zoom_out)
plt.title('Zoom out')
```
