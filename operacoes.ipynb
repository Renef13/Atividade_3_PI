```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
# Carregar a imagem
imagem = cv2.imread('data/pirata.jpeg')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
plt.title('Original')
print(f'O pixel minimo é {imagem_rgb.min()} e o máximo é {imagem_rgb.max()}')
print(f'O formato da imagem é {imagem_rgb.shape}')
```
