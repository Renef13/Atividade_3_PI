```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carregar a imagem
imagem = cv2.imread('data/pirata.jpeg')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
plt.title('Original')
altura = imagem_rgb.shape[0]
largura = imagem_rgb.shape[1]
canais_cor = imagem_rgb.shape[2]
print(f'A largura da imagem é {largura}, a altura é {altura} e ela tem {canais_cor} canais de cor')
```

```python
#Transladar a imagem
tx, ty = 100, 50
M_translacao = np.float32([[1, 0, tx], [0, 1, ty]])
imagem_transladada = cv2.warpAffine(imagem_rgb, M_translacao, (largura, altura))
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.title('Transladada')
plt.imshow(imagem_transladada)
```

```python
# ZoomIn e ZoomOut

InSx, InSy = 1.5, 1.5
OutSx, OutSy = 0.5, 0.5

M_In = np.float32([[InSx, 0, 0], [0, InSy, 0]])
M_Out = np.float32([[OutSx, 0, 0], [0, OutSy, 0]])

imagem_zoom_in = cv2.warpAffine(imagem_rgb, M_In, (largura, altura))
imagem_zoom_out = cv2.warpAffine(imagem_rgb, M_Out, (largura, altura))
plt.subplot(1, 2, 1)
plt.imshow(imagem_zoom_in)
plt.title('Zoom in')
plt.subplot(1, 2, 2)
plt.imshow(imagem_zoom_out)
plt.title('Zoom out')
```

```python
#Zoom In e Zoom Out com cv2.resize
imagem_zoom_in = cv2.resize(imagem_rgb, None, fx=InSx, fy=InSy)
imagem_zoom_out = cv2.resize(imagem_rgb, None, fx=OutSx, fy=OutSy)
plt.subplot(1, 2, 1)
plt.imshow(imagem_zoom_in)
plt.title('Zoom in')
plt.subplot(1, 2, 2)
plt.imshow(imagem_zoom_out)
plt.title('Zoom out')
```

```python
#Rotação

angulo = -180
theta = np.radians(angulo)
cos = np.cos(theta)
sin = np.sin(theta)

cx = largura / 2
cy = altura / 2

tx = (1 - cos) * cx - sin * cy
ty = sin * cx + (1 - cos) * cy
M_rotacao = np.float32([[cos, -sin, tx], [sin, cos, ty]])

imagem_rotacionada = cv2.warpAffine(imagem_rgb, M_rotacao, (largura, altura))
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(imagem_rotacionada)
plt.title('Rotacionada')
# Falta corrigir pra nao sair da dela
```

```python
# Rotação com cv2.getRotationMatrix2D

angulo = -25
M_rotacao = cv2.getRotationMatrix2D((largura, altura), angulo, 0.5)

imagem_rotacionada = cv2.warpAffine(imagem_rgb, M_rotacao, (largura, altura))
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(imagem_rotacionada)
plt.title('Rotacionada')
```

```python
# Espelhamento horizontal

M_reflexao_horizontal = np.float32([[-1, 0, largura], [0, 1, 0]])
imagem_refletida_horizontal = cv2.warpAffine(imagem_rgb, M_reflexao_horizontal, (largura, altura))
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(imagem_refletida_horizontal)
plt.title('Espelhada Horizontal')
```

```python
# Espelhament horiontal com cv2.flip

imagem_refletida_horizontal = cv2.flip(imagem_rgb, 1)
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(imagem_refletida_horizontal)
plt.title('Espelhada Horizontal')
```

```python
# Espelhamento vertical

M_reflexao_vertical = np.float32([[1, 0, 0], [0, -1, altura]])
imagem_refletida_vertical = cv2.warpAffine(imagem_rgb, M_reflexao_vertical, (largura, altura))
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(imagem_refletida_vertical)
plt.title('Espelhada Vertical')
```

```python
# Espelhamento vertical com cv2.flip

imagem_refletida_vertical = cv2.flip(imagem_rgb, 0)
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(imagem_refletida_vertical)
plt.title('Espelhada Vertical')
```

```python
# Cisalhamento

sh_x = 0.5
sh_y = 0.5

M_cisalhamento = np.float32([[1, sh_x, 0], [sh_y, 1, 0]])

nova_largura = int(largura + abs(sh_x) * altura)
nova_altura = int(altura + abs(sh_y) * largura)
imagem_cisalhada = cv2.warpAffine(imagem_rgb, M_cisalhamento, (nova_largura, nova_altura))
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(imagem_cisalhada)
plt.title('Cisalhamento')
```

```python
#converter as matrizes pro tamanho certo
def mat_3x3(M):
   return np.vstack([M, [0, 0, 1]]).astype(np.float32)
```

```python
# Combinação de rotação, reflexão e cisalhamento

angulo = -180
theta = np.radians(angulo)
cos = np.cos(theta)
sin = np.sin(theta)

cx = largura / 2
cy = altura / 2

tx = (1 - cos) * cx - sin * cy
ty = sin * cx + (1 - cos) * cy
M_rotacao = np.float32([[cos, -sin, tx], [sin, cos, ty]])

sh_x = 0.5
sh_y = 0.5

M_cisalhamento = np.float32([[1, sh_x, 0], [sh_y, 1, 0]])

nova_largura = int(largura + abs(sh_x) * altura)
nova_altura = int(altura + abs(sh_y) * largura)

M_reflexao_horizontal = np.float32([[-1, 0, largura], [0, 1, 0]])

mat_rotacao = mat_3x3(M_rotacao)
mat_cisalhamento = mat_3x3(M_cisalhamento)
mat_reflexao_horizontal = mat_3x3(M_reflexao_horizontal)

matriz_combinada = mat_cisalhamento @ mat_reflexao_horizontal @ mat_rotacao

matriz_resultado = matriz_combinada[:2,:]

imagem_combinada = cv2.warpAffine(imagem_rgb, matriz_resultado, (nova_largura, nova_altura))
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(imagem_combinada)
plt.title('Combinada')
```
